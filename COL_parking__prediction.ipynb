```python
import datetime

# Create a new column called "day_of_week".
datak["day_of_week"] = datak["parking_start_date"].dt.weekday

# Filter the data so that only the days from 1 to 7 are included.
datak = datak[datak["day_of_week"].isin([0, 1, 2, 3, 4, 5, 6])]

# Print the DataFrame.
(datak.head(2))
```

```python
# Map the days of the week to their actual names.
day_of_week_map = {
    0: "Monday",
    1: "Tuesday",
    2: "Wednesday",
    3: "Thursday",
    4: "Friday",
    5: "Saturday",
    6: "Sunday"
}
datak["day_of_week"] = datak["day_of_week"].apply(lambda day_of_week: day_of_week_map[day_of_week])

# Inspect the DataFrame to make sure that the day_of_week column is correct.
print(datak.head(2))
```

```python
datak['day_type'] = datak['day_of_week'].apply(lambda x: 'Weekend' if x in ['Saturday','Sunday'] else 'Weekday')
```

```python
datak['parking_hour'] = datak['parking_start_date'].dt.hour
datak['time_of_day'] = datak['parking_hour'].apply(lambda x: 'Morning' if x < 12 else ('Afternoon' if x < 18 else 'Evening'))
```

```python
import pandas as pd
import datetime

def _get_season(parking_start_date):
    month = parking_start_date.month
    if month in [12, 1, 2]:
        return "Winter"
    elif month in [3, 4, 5]:
        return "Spring"
    elif month in [6, 7, 8]:
        return "Summer"
    else:
        return "Fall"

# Create a new column for seasons.
datak["season"] = datak["parking_start_date"].apply(lambda parking_start_date: _get_season(parking_start_date))

# Inspect the DataFrame to make sure that the season column is correct.
datak.head(2)
```

```python
import pandas as pd
import holidays
```

```python
# Fetch public holidays for Ohio, USA
ohio_holidays = holidays.US(years=2019, state='OH')
```

```python
# Create a new column called "public_holiday".
datak["public_holiday"] = 0
```

```python
# Fill the new column with the public holidays for the Aus.
for index, row in datak.iterrows():
    date = row['parking_start_date']
    if date in ohio_holidays:
        datak.loc[index, "public_holiday"] = 1
```
